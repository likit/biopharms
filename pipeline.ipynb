```python
import networkx as nx
from collections import defaultdict
```

```python
author_graph = nx.DiGraph()
for article in db.scopus.find():
    for author in article['creator']:
        for coauthor in article['creator']:
            if coauthor != author:
                if author not in author_graph:
                    author_graph.add_edge(author, coauthor)
                    author_graph[author][coauthor]['copub'] = 1
                else:
                    if coauthor not in author_graph[author]:
                        author_graph.add_edge(author, coauthor)
                        author_graph[author][coauthor]['copub'] = 1
                    else:
                        author_graph[author][coauthor]['copub'] += 1
```

```python
copubs = defaultdict(int)
for edge in author_graph.edges_iter():
    author, coauthor = edge
    copub = author_graph[author][coauthor]['copub']
    if copub > 1:
        print author, coauthor, copub
    copubs.append(copub)
```

```python
for edge in author_graph.edges_iter():
    author, coauthor = edge
    copub = author_graph[author][coauthor]['copub']
    if copub == max(copubs):
        print author, coauthor, copub
```

```python
%matplotlib inline
```

```python
G = nx.DiGraph()
author = 'Sattabongkot, Jetsumon'
for coauthor in author_graph[author]:
    copub = author_graph[author][coauthor]['copub']
    if copub > 1:
        G.add_edge(author, coauthor)
        G[author][coauthor] = author_graph[author][coauthor]  # copub
        print author, coauthor, copub
```

```python
copubs = []
edge_labels = dict()
for (u,v,d) in G.edges(data=True):
    print u,v,d
    copubs.append(d['copub'])
    edge_labels[(u,v)] = d['copub']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
pos = nx.spring_layout(G)
nx.draw_networkx_edges(G, pos, alpha=0.3, width=copubs, edge_color='b', arrows=False)
nx.draw_networkx_nodes(G, pos, node_size=[500, 700, 800, 100, 200, 300, 450, 500],
                      node_color='m', alpha=0.7)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=15)
#nx.draw_spring(G)
plt.savefig('author-graph-sample3.png', dpi=100)
```
